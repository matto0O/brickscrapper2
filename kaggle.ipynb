```python
path = "sets_filtered_full.csv"

import pandas as pd

def load_data(path: str = path) -> pd.DataFrame:
    return pd.read_csv(path)

df = load_data()

del df['theme']
del df['img_url']

df['_id'] = df["set_id"]

del df['set_id']

df['sim_scores'] = df['sim_scores'].apply(lambda x: list(sorted(eval(x), key=lambda y: y[1], reverse=True)))
```

```python
df.to_json('sets_filtered_full.json', orient='records', lines=True)
```

```python
parts_all = set()

for index, row in df.iterrows():
    parts = eval(row['parts'])
    for key in parts.keys():
        parts_all.add(key[0])
```

```python
len(parts_all)
```

```python
import json
json.dump(list(parts_all), open(f"parts.json", "w"))
```

```python
from brick_scanner import PartScanner
from utils import split_workload
from collections import deque

THREAD_COUNT = 4
workloads = split_workload(list(parts_all), THREAD_COUNT)

threads = []

for i in range(THREAD_COUNT):
    q = deque(workloads[i])
    scanner = PartScanner(index=i, headless=True, data_source=q)
    scanner.start()
    threads.append(scanner)

for scanner in threads:
    scanner.join()
```
